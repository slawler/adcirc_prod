### 1. Load the libraries and the stations

```python
from ipywidgets import *
import matplotlib.pyplot as plt
import pandas as pd
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True)
from adcirc import adcirc
from importlib import reload
import noaa_lib ; reload(noaa_lib)
from noaa_lib import *
stations = tide_gauges()
```

### 2. Create Map with the Station Locations

```python
lat1,lat2 = 34.1, 35.3
lon1,lon2 = -78, -75.6
fig,ax=plt.subplots(figsize=(48,8))
adcirc.noaa_map(ax,lat1,lat2,lon1,lon2)
```

### 3. Create Hurricane Time Series Plots

```python
begin,end = '20160930','20161014'
title = 'Hurricane Matthew'
data,layout = adcirc.plot_hurricane(begin,end,stations,title)
fig = go.Figure(data=data, layout=layout)
interactive = iplot(fig)
```
