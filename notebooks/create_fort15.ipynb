### 3. Read the fort template and replace variables that were in the csv

```python
with open(fort15,'w') as f_in:
    with open(templ,'r') as f_out:
        lines = f_out.readlines()
        for i, line in enumerate(lines):
            for ii in range(0,len(table)):
                if table['variable'][ii] in line:
                    new_v = line.split(' ')[0]
                    line = line.replace(new_v,str(table['defin'][ii]+'\t'))
            f_in.write(line)
```
